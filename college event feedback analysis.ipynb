```python
# downloading NLTK data files
import nltk
nltk.download('punkt')
nltk.download('vader_lexicon')
```

```python
# using VADER sentiment analysis tool
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
# Ensure all entries are strings
students['Cleaned Questions'] = students['Cleaned Questions'].apply(
    lambda x: ' '.join(x) if isinstance(x, list) else str(x)
)

# Now apply sentiment analysis
students['Sentiment'] = students['Cleaned Questions'].apply(
    lambda x: sia.polarity_scores(x)['compound']
)

# Categorize sentiment
students['Sentiment Category'] = students['Sentiment'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)
```

```python
students
```

```python
# sentiment analysis using textblob
from textblob import TextBlob
# function to get sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity
# applying sentiment analysis to the 'Questions' column
students['TextBlob Sentiment'] = students['Cleaned Questions'].apply(get_sentiment)
# categorizing sentiment scores into positive, begative, and neutral
students['TextBlob Sentiment Category'] = students['TextBlob Sentiment'].apply(lambda x:
    'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
# displaying the sentiment analysis results
textblob_results = students[['Cleaned Questions', 'TextBlob Sentiment', 'TextBlob Sentiment Category']]
textblob_results
```

```python
# comparing accuracy between VADER and TextBlob sentiment analysis
accuracy_comparison = students[['Cleaned Questions', 'Sentiment Category', 'TextBlob Sentiment Category']]
accuracy_comparison['Match'] = accuracy_comparison['Sentiment Category'] == accuracy_comparison['TextBlob Sentiment Category']
accuracy_comparison['Match'] = accuracy_comparison['Match'].map({True: 'Match', False: 'Mismatch'})
accuracy_comparison
```

```python
# identifying the most accurate sentiment analysis method
vader_accuracy = accuracy_comparison['Match'].value_counts().get('Match', 0)
textblob_accuracy = accuracy_comparison['Match'].value_counts().get('Mismatch', 0)
print(f"VADER Sentiment Analysis Accuracy: {vader_accuracy}")
print(f"TextBlob Sentiment Analysis Accuracy: {textblob_accuracy}")
# --- IGNORE ---
```

```python
# testing the accuracy between using vader and textblob
from sklearn.metrics import accuracy_score
# mapping sentiment categories to numerical values for comparison
sentiment_mapping = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
vader_sentiments = students['Sentiment Category'].map(sentiment_mapping)
textblob_sentiments = students['TextBlob Sentiment Category'].map(sentiment_mapping)
accuracy = accuracy_score(vader_sentiments, textblob_sentiments)
print(f"Accuracy between VADER and TextBlob sentiment analysis: {accuracy:.2f}")
# visualizing the results
import matplotlib.pyplot as plt
import seaborn as sns
# setting the style for seaborn
sns.set(style="whitegrid")
# plotting the sentiment distribution
plt.figure(figsize=(10,6))
sns.countplot(data=students, x='Sentiment Category', palette='viridis')
plt.title('Sentiment Distribution of Feedback Questions')
plt.xlabel('Sentiment Categoory')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# identifying the most common words in each sentiment category
from collections import Counter
# function to extract words from a text
def extract_words(text):
    words = text.split()
    return [word.lower() for word in words if word.isalpha()]
# extracting words from the 'Questions' column
students['Words'] = students['Cleaned Questions'].apply(extract_words)
# counting the frequency of words in each sentiment category
word_counts = students.explode('Words').groupby('Sentiment Category')['Words'].apply(lambda x: Counter(x).most_common(10)).reset_index()
word_counts
# visualizing the most common words in each sentiment category
plt.figure(figsize=(12,8))
for sentiment in word_counts['Sentiment Category'].unique():
    words = word_counts[word_counts['Sentiment Category'] == sentiment]['Words'].values[0]
    counts = [count for word, count in words]
    labels = [word for word, count in words]
    plt.bar(labels, counts, label=sentiment)
    plt.title(f'Most Common Words in {sentiment} Sentiment Category')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
